# **Lab 1: The ratio of noun to verbs in reviewtext using the pos tag**

```
import pandas as pd 
import spacy
```

```
data=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines= True,orient='values')
```

```
data.head()
```

```
model = spacy.load("en_core_web_sm")
```

```
def nounverb(data1):
  noun=0
  verb=0
  words=model(data1)
  for i in words:
    if i.pos_ =='NOUN':
      noun+=1
    elif i.pos_ =='VERB':
      verb+=1
  if noun!=0 and verb !=0:
    ratio=noun/verb
  else:
    ratio=0
  answer = {'Number_of_nouns' : noun,'Number_of_verbs' : verb,'Noun:Verb' : ratio}
  return answer
```

```
data['Ratio Noun to Verb']=data['reviewText'].apply(nounverb)
```
