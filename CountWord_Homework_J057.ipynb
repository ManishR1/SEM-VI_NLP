```
import numpy as np
import pandas as pd
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines = True)
```

```
data.head()
```

```
def countword(str):
  count = dict()
  words = str.split()
  for i in words:
    if i in count:
      count[i]+=1
    else:
      count[i]=1
  return count
```

```
data['wordcount_reviewText'] = data['reviewText'].apply(countword)
```

```
data['wordcount_reviewText']
```

```
data['wordcount_summary'] = data['summary'].apply(countword)
```
