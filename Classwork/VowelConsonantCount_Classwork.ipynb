#**Class 2: Count of Vowels and Consonants**

```
import pandas as pd
import re
```

```
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz", lines=True)
```

```
data.head()
```

```
def vowcons(data):
    data1 = data['reviewText']
    data1 = data1.lower()
    data1 = re.sub(r'[^a-z0-9\s]', '', data1)
    vowel_count=0
    consonant_count=0
    for letter in data1:
        if letter in ['a', 'e', 'i', 'o', 'u']:
            vowel_count+=1
        elif letter.isalpha():
            consonant_count+=1
    data1 = data1.split(' ')
    return {'vowel_count':vowel_count,'consonant_count': consonant_count}
```

```
data['reviewText_VowConsCount'] = data.apply(vowcons, axis=1)
```
