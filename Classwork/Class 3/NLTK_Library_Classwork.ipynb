```
import pandas as pd
import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```
lem=WordNetLemmatizer()
```

```
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz', lines=True)
```

```
data.head()
```

```
def nltk_lib(data1):
  final=dict()
  token = nltk.word_tokenize(data1)
  final['tokenization'] = nltk.word_tokenize(data1)
  final['pos_tag'] = nltk.pos_tag(data1.split())
  final['lemmatization']=[lem.lemmatize(word) for word in token]
  return final
```

```
def nltk_li(data1):
  final=dict()
  final['tokenization'] = nltk.word_tokenize(data1)
  return final
```

```
data['reviewText'].apply(nltk_li)
```

```
def nltk2(data1):
  final=dict()
  token = nltk.word_tokenize(data1)
  final['lemmatization']=[lem.lemmatize(word) for word in token]
  return final
```

```
data['reviewText'].apply(nltk2)
```

```
words=dict(nltk.pos_tag(data['reviewText'][0].split()))
```

```
adj=[i for i in words if words[i]=='JJ']
```
