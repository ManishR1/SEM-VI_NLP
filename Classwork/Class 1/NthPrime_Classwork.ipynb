#**Class 1: Finding the Nth prime number**

```
##function for finding the nth prime number
def nthprime(n):
  if n==1 :
    return 2
  count = 1
  s = 1
  while(count < n):
    s += 2
    if prime(s):
      count += 1
  return s
```

```
##function to find if the number is prime or not
def prime(num):
  for i in range (2,num//2):
    if (num % i == 0):
      return False
  return True
```
